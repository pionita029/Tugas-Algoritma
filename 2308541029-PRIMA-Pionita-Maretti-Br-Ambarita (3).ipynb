```python
 # Diketahui bahwa ùëõ adalah dua digit terakhir dari NIM, maka tentukan biangan prima antara 1000ùëõ¬†dan¬†1000(ùëõ+1).
 # Dua digit terakhir nim saya adalah 29
 # n = 29. 1000(29) dan 1000(29+1)
 # menentukan bilangan prima antara 29000 dan 30000
print("Di sini kita akan menentukan bilangan prima antara 29000 hingga 30000")
nilai_awal_29 = 29000
nilai_akhir_29 = 30000

print("bilangan prima mulai dari",nilai_awal_29 , "hingga",nilai_akhir_29 , " ")
for i_29 in range(nilai_awal_29, nilai_akhir_29 + 1):
    count = 0
    for n_29 in range(2, i_29):
        if i_29 % n_29 == 0:
            count = 1
            break
    if count == 0:
        print(i_29, end = " ")
    
        
```
