```python
print("disini kita akan mencari persamaan garis yang memuat tiga kasus")

# Kita akan mencari titik potong dari dua garis yg memenuhi salah satu dari ketiga kasus di atas
import numpy as np
import matplotlib.pyplot as plt

kasus = float(input("Kasus yang akan dikerjakan adalah kasus "))

if kasus==1:
    print("Masing-masing garis akan melewati dua titik yang diketahui.")
    i_29 = float(input("Nilai x dari titik pertama dari garis pertama:  "))
    j_29 = float(input("Nilai y dari titik pertama dari garis pertama: "))
    k_29 = float(input("Nilai x dari titik kedua dari garis pertama:  "))
    l_29 = float(input("Nilai y dari titik kedua dari garis pertama: "))

    m_29 = float(input("Nilai x dari titik pertama dari garis kedua:  "))
    n_29 = float(input("Nilai y dari titik pertama dari garis kedua: "))
    o_29 = float(input("Nilai x dari titik kedua dari garis kedua: "))
    p_29 = float(input("Nilai y dari titik kedua dari garis kedua: "))
    
    m_1 = (l_29-j_29)/(k_29-i_29) # Gradien garis pertama
    m_2 = (p_29-n_29)/(o_29-m_29) # Gradien garis kedua

    # Persamaan garis pertama adalah y=m_1(x-i_29)+j_29=m_1x-m_1i_29+j_29
    # Persamaan garis kedua adalah y=m_2(x-m_29)+f=m_2x-m_2m_29+n_29

    c_1 = -m_1*i_29+j_29 # Titik potong garis pertama terhadap sumbu y
    c_2 = -m_2*m_29+n_29 # titik potong garis kedua terhadap sumbu y garis kedua

    print("Persamaan garis pertama y={:.3f}x+{:.3f}".format(m_1,c_1))
    print("Persamaan garis kedua y={:.3f}x+{:.3f}".format(m_2,c_2))

    if m_1==m_2:
    #print("Kedua garis sejajar, kemungkinannya tidak terdapat solusi atau terdapat tak berhingga solusi")
        if c_1 == c_2:
            print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga  solusi")
            x=np.linspace(-5,5,100)
            plt.plot(x,m_1*(x-i_29)+j_29,color='red')
            plt.plot(x,m_2*(x-m_29)+n_29,color='blue')
            plt.axvline(x=0,color='grey')
            plt.axhline(y=0,color='grey')
            plt.show()
        else:
            print("Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.")
            x=np.linspace(-5,5,100)
            plt.plot(x,m_1*(x-i_29)+j_29,color='red')
            plt.plot(x,m_2*(x-m_29)+n_29,color='blue')
            plt.axvline(x=0,color='grey')
            plt.show()
    else:
        print("Akan terdapat satu titik potong")
        tp_x = (-m_2*m_29+n_29+m_1*i_29-j_29)/(m_1-m_2)
        tp_y = m_1*(tp_x-i_29)+j_29
        print("Titik potong kedua garis adalah ({:.3f},{:.3f})".format(tp_x,tp_y))
    
        x=np.linspace(tp_x-5,tp_x+5,100)
        plt.plot(x,m_1*(x-a)+b,color='red',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m_1,i_29,j_29))
        plt.plot(x,m_2*(x-e)+f,color='blue',label="y={:.3f}(x-{:.3f})+{:.3f}".format(m_2,m_29,n_29))
        plt.axvline(x=0,color='grey')
        plt.legend(loc='best')
        plt.show()
        
elif kasus==2:
    print("Masing-masing garis diketahui gradien dan titik yang dilewati.")
    m_1 = float(input("Gradien dari garis pertama "))
    i_29 = float(input("Nilai x dari titik yang dilewati oleh garis pertama "))
    j_29 = float(input("Nilai y dari titik yang dilewati oleh garis pertama "))
    print("Persamaan garis pertama y={}(x-{})+{}".format(m_1,i_29,j_29))
    
    m_2 = float(input("Gradien dari garis kedua "))
    k_29 = float(input("Nilai x dari titik yang dilewati oleh garis kedua "))
    l_29 = float(input("Nilai y dari titik yang dilewati oleh garis kedua "))
    print("Persamaan garis kedua y={}(x-{})+{}".format(m_2,k_29,l_29))
    
    c_1 = -m_1*i_29+j_29
    c_2 = -m_2*k_29+l_29
    
    if m_1==m_2:
        print("Kedua garis sejajar")
        if c_1 == c_2:
            print("Terdapat tak berhingga solusi")
        else:
            print("Tidak terdapat solusi")
    else:
        print("Kedua garis akan berpotongan di satu titik.")
        
elif kasus==3:
    print("Kedua garis memenuhi persamaan berbentuk ax+by=c dan dx+ey=f.")
    a=float(input("Koefisien dari x pada garis pertama adalah "))
    b=float(input("Koefisien dari y pada garis pertama adalah "))
    c=float(input("Konstanta dari garis pertama adalah "))
    
    d=float(input("Koefisien dari x pada garis kedua adalah "))
    e=float(input("Koefisien dari y pada garis kedua adalah "))
    f=float(input("Konstanta dari garis kedua adalah "))

    m_1 = (-i_29)/j_29
    m_2 = (-k_29)/l_29
    
    c_1=k_29/j_29
    c_2=n_29/m_29
    
    if m_1==m_2:
        print("Kedua garis akan sejajar")
        if c_1==c_2:
            print("Kedua garis berhimpitan")
        else:
            print("Kedua garis sejajar")
    
    else:
        print("kedua garis akan berpotongan di satu titik")
else:
    print("Hanya terdapat 3 kasus. Mohon masukkan 1, 2 atau 3 saja.")
    
```
