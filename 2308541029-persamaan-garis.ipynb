```python
#Mencari gradien garis
import numpy as np
import matplotlib.pyplot as plt

#Titik (a_29,b_29)
a_29 = 1
b_29 = 3

#Titik (c_29,d_29)
c_29 = 3
d_29 = 5

# memberi label untuk x dan y
x_29 = plt.xlabel('Sumbu X')
y_29 = plt.ylabel('Sumbu Y')

#Menghitung gradien
m_29 =(d_29-b_29)/(c_29-a_29)
print("Gradien garisnya adalah {}".format(m_29))
x_29 = np.arange(-2,4)
plt.plot(x_29,m_29*(x_29-a_29)+b_29)

#Membuat kotak dalam plot
plt.grid()
#Menampilkan hasil
plt.show()



```

```python
#Mencari Persamaan Garis Saling Berpotongan
import numpy as np
import matplotlib.pyplot as plt

#Persamaan Garis Melalui Titik (e,f) dan (g,h)
#Titik (e_29,f_29). Misalkan e=x1, f=y1
e_29=2 
f_29=3
#Titik (g_29,h_29). Misalkan g=x2, h=y2
g_29=3 
h_29=2
m1=(h_29-f_29)/(g_29-e_29)
print("Gradien garis pertama adalah {}".format(m1))
print("Persamaan garis pertama adalah y={}x+{} ".format(m1,-m1*e_29+f_29))


# Persamaan garis melewati titik (i,j) dan (k,l)
#Titik (i_29,j_29). Misalkan i=x3, j=y3
i_29=3
j_29=1
#Titik (k_29,l_29). Misalkan k=x4, l=y4
k_29=4 
l_29=-8
m2=(l_29-j_29)/(k_29-i_29)
print("Gradien garis ke dua adalah {}".format(m2))
print("Persamaan garis kedua adalah y={}x+{} ".format(m2,-m2*i_29+j_29))


if m1==m2:
    print("kedua gradien sama, kemungkinan garisnya sejajar atau berhimpit.")
else:
    # Titik potong kedua garis pada (tp_x,tp_y)
    tp_x = (m1*e_29-f_29-m2*i_29+j_29)/(m1-m2)
    tp_y = (m2*(tp_x-i_29)+j_29)
    print("Titik potongnya adalah ({},{})".format(tp_x,tp_y))

    x=np.arange(tp_x-5,tp_x+5,0.01)
    plt.figure(figsize=(5,5), dpi=100)
    
    plt.plot(x,m1*(x-e_29)+f_29,color='yellow',label='y={}x+{}'.format(m1,-m1*e_29+f_29))
    plt.plot(x,m2*(x-i_29)+j_29,color='violet',label='y={}x+{}'.format(m2,-m2*i_29+j_29))
    plt.plot(tp_x,tp_y,color='green',marker='o',markersize=3)
    plt.legend(loc='best')
    
    #Membuat kotak dalam plot
    plt.grid()
    #Menampilkan hasil
    plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

#Persamaan Garis Melalui Titik (a,b) dan (c,d)
#Titik (a_29,b_29). Misalkan a=x1, b=y1
a_29 = 6
b_29 = 4
#Titik (c_29,d_29). Misalkan c=x2, d=y2
c_29 = 5
d_29 = 3

#Persamaan Garis Melalui Titik (e,f) dan (g,h)
#Titik (e_29,f_29). Misalkan e=x3, b=y3
e_29 = 3
f_29 = 5
#Titik (g_29,h_29). Misalkan g=x4, h=y4
g_29 = 4
h_29 = 6

#Persamaan Garis Melalui Titik (i,j) dan (k,l)
#Titik (i_29,j_29). Misalkan i=x5, j=y5
i_29 = 4
j_29 = 5
#Titik (k_29,l_29). Misalkan a=x6, b=y6
k_29 = 3
l_29 = 6

#Menentukan gradien
m1 =(d_29-c_29)/(b_29-a_29) #gradien 1
m2 =(h_29-g_29)/(f_29-e_29) #gradien 2
m3 =(l_29-k_29)/(j_29-i_29) #gradien 3

x = np.arange(-3,5)

plt.plot(x,m1*(x-a_29)+b_29) 
plt.plot(x,m2*(x-e_29)+f_29)
plt.plot(x,m3*(x-i_29)+j_29)

y1 = m1*(x-a_29)+b_29
y2 = m2*(x-e_29)+f_29
y3 = m3*(x-i_29)+j_29

# memberi label untuk x dan y
x_29 = plt.xlabel('Sumbu X')
y_29 = plt.ylabel('Sumbu Y')

#Gambar grafik
plt.plot(x,m1*(x-a_29)+b_29,color='red',label='y={}x+{}'.format(m1,-m1*a_29+b_29))
plt.plot(x,m2*(x-e_29)+f_29,color='blue',label='y={}x+{}'.format(m2,-m2*e_29+f_29))
plt.plot(x,m3*(x-i_29)+f_29,color='green',label='y={}x+{}'.format(m3,-m3*i_29+j_29))

#hasil
plt.title("GAMBAR GRAFIK")
print("Akan diperoleh bentuk grafik seperti dibawah ini!")
plt.show()
```
